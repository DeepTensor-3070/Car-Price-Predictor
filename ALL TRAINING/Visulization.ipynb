```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Cleaned_Data.csv")
```

```python
data
```

```python
data.info()
```

```python
data['company'].unique()
```

Company VS Price

```python

ax=sns.boxplot(x="company",y="Price",data = data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
```

Year VS Price

```python
ax=sns.boxplot(x="year",y="Price",data = data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
plt.title('Year VS Price')
```

kms_driven VS Price

```python
sns.relplot(x='kms_driven',y='Price',data=data,height=7,aspect=1.5)
```

Fuel_Type VS Price 

```python
sns.boxplot(x='fuel_type',y='Price',data=data)
```

Price with Fueltype,Year and Comapny

```python
ax=sns.relplot(x='company',y='Price',data=data,hue='fuel_type',size='year',height=7,aspect=2)
ax.set_xticklabels(rotation=40,ha='right')
```
